```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def criar_grafico_comparativo(x1,y1,x2,y2,titulo,intervalo_x = None, intervalo_y = None,tempo = '(S)'):    
    plt.figure(figsize=(16,9))
    plt.gca().set_facecolor('#ffffff')
    plt.gcf().set_facecolor('#be7af3')
    plt.title(titulo, font = {'size':17})
    plt.grid()
    plt.plot(x1,y1, marker='o', color='#070A40', label='Quick sort')
    plt.plot(x2,y2, marker='o', color='#69AFBF', label = 'Insertion sort')
    if intervalo_x != None:
        plt.xticks(intervalo_x)
    if intervalo_y != None:
        plt.yticks(intervalo_y)
    plt.legend()
    plt.ylabel('Tempo ' + tempo, font = {'size':14})
    plt.xlabel('Quantidade de Termos', font = {'size':14})
    nome_img = titulo + '.svg'
    plt.savefig(nome_img, format='svg')
    plt.show()
```

```python
def criar_grafico_individual(x1,y1,y2,y3,titulo,intervalo_x = None, intervalo_y = None,tempo = '(S)'):    
    plt.figure(figsize=(16,4.5))
    plt.gca().set_facecolor('#ffffff')
    plt.gcf().set_facecolor('#be7af3')
    plt.title(titulo, font = {'size':17})
    plt.grid()
    plt.plot(x1,y1, marker='o', color='#070A40', label='Lista Aleat√≥ria')
    plt.plot(x1,y2, marker='o', color='#69AFBF', label = 'Lista Ordenada')
    plt.plot(x1,y3, marker='o', color='#F9DB82', label = 'Lista Ordenada Reversa')
    if intervalo_x != None:
        plt.xticks(intervalo_x)
    if intervalo_y != None:
        plt.yticks(intervalo_y)
    plt.legend()
    plt.ylabel('Tempo ' + tempo, font = {'size':14})
    plt.xlabel('Quantidade de Termos', font = {'size':14})
    nome_img = titulo + '.svg'
    plt.savefig(nome_img, format='svg')
    plt.show()
```
