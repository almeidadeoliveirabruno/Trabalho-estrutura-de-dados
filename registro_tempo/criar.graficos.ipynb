```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

# Listas pequenas

```python
tempos_pequenos_insertion = pd.read_csv('tempos_pequenas_insertion_sort.csv',encoding='utf-8')
tempos_pequenos_insertion.head()
```

## Lista Pequena aleatoria

```python
insertion_lista_pequena_aleatoria = tempos_pequenos_insertion.iloc[0:10,:]
insertion_lista_pequena_aleatoria.head(11)
```

```python
tempos_pequenos_quicksort = pd.read_csv('tempos_pequenas_Quick_sort.csv')
tempos_pequenos_quicksort.head(5)
```

```python
quicksort_lista_pequena_aleatoria =  tempos_pequenos_quicksort.iloc[0:10,:]
quicksort_lista_pequena_aleatoria.head(11)
```

```python
def criar_grafico(x1,y1,x2,y2,titulo,intervalo_x = None, intervalo_y = None,tempo = '(S)'):    
    plt.figure(figsize=(16,9))
    plt.gca().set_facecolor('#F2F2F2')
    plt.gcf().set_facecolor('#9AC7D9')
    plt.title(titulo, font = {'size':17})
    plt.grid()
    plt.plot(x1,y1, marker='o', color='#070A40', label='Quick sort')
    plt.plot(x2,y2, marker='o', color='#69AFBF', label = 'Insertion sort')
    if intervalo_x != None:
        plt.xticks(intervalo_x)
    if intervalo_y != None:
        plt.yticks(intervalo_y)
    plt.legend()
    plt.ylabel('Tempo ' + tempo, font = {'size':14})
    plt.xlabel('Quantidade de Termos', font = {'size':14})
    #plt.savefig('Gr√°fico Exponencial.png', dpi=300)
    plt.show()
```

```python
x1 = quicksort_lista_pequena_aleatoria['Tamanho'].values
y1 = quicksort_lista_pequena_aleatoria['Tempo (ns)'].values
x2 = insertion_lista_pequena_aleatoria['Tamanho'].values
y2 = insertion_lista_pequena_aleatoria['Tempo (ns)'].values
intervalo_x = range(10,101,10)
criar_grafico(x1,y1,x2,y2,'Lista Pequena Aleat√≥ria',intervalo_x,tempo='(NS)')
```

## Lista Pequena ordenada

```python
quicksort_lista_pequena_ordenada =  tempos_pequenos_quicksort.iloc[10:20,:]
quicksort_lista_pequena_ordenada.head(11)
```

```python
insertion_lista_pequena_ordenada = tempos_pequenos_insertion.iloc[10:20,:]
insertion_lista_pequena_ordenada.head(11)
```

```python
x1 = quicksort_lista_pequena_ordenada['Tamanho'].values
y1 = quicksort_lista_pequena_ordenada['Tempo (ns)'].values
x2 = insertion_lista_pequena_ordenada['Tamanho'].values
y2 = insertion_lista_pequena_ordenada['Tempo (ns)'].values
intervalo_x = range(10,101,10)
criar_grafico(x1,y1,x2,y2,'Lista Pequena ordenada',intervalo_x, tempo='(NS)')
```
